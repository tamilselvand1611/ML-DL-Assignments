```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.discrete.count_model import ZeroInflatedPoisson
```

```python
df=pd.read_csv("insurance2.csv")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df[["age", "bmi", "charges", "insuranceclaim"]]
df
```

```python
df.isnull().sum()
```

```python
X=df[["age", "bmi", "charges"]]
y=df["insuranceclaim"]
```

```python
X=sm.add_constant(X)
X.head()
```

```python
model=ZeroInflatedPoisson(endog=y, exog=X, exog_infl=X)
```

```python
result=model.fit(maxiter=200, method="bfgs")
```

```python
print(result.summary())
```

```python
print(df["insuranceclaim"].value_counts())
```

```python
zero_percentage=((df["insuranceclaim"]==0).mean())*100
print("Zero Percentage:", round(zero_percentage,2),"%")
```

```python
age = int(input("Enter Age: "))
bmi = float(input("Enter BMI: "))
charges = float(input("Enter Charges: "))
```

```python
new_data = pd.DataFrame({"const":[1],"age":[age],"bmi":[bmi],"charges":[charges]})
```

```python
prediction = result.predict(exog=new_data,exog_infl=new_data)
print("Predicted Insurance Claim:",round(prediction[0]))
```
