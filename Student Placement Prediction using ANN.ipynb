```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
data=pd.read_csv("placementdata.csv")
data
```

```python
data.columns
```

```python
data["PlacementStatus"].value_counts()
```

```python
le=LabelEncoder()
data["ExtracurricularActivities"]=le.fit_transform(data["ExtracurricularActivities"])
data["PlacementTraining"]=le.fit_transform(data["PlacementTraining"])
```

```python
X=data[["CGPA", "Internships", "Projects", "Workshops/Certifications", "AptitudeTestScore", "SoftSkillsRating", "ExtracurricularActivities", "PlacementTraining", "SSC_Marks", "HSC_Marks"]]
X.head()
```

```python
scaler=StandardScaler()
X=scaler.fit_transform(X)
```

```python
y=data["PlacementStatus"]
y.head()
print(y.shape)
```

```python
y=le.fit_transform(y)
print(le.classes_)
```

```python
model=Sequential()
model.add(Dense(16, activation="relu"))
model.add(Dense(8, activation="relu"))
model.add(Dense(1, activation="sigmoid"))
```

```python
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
model.fit(X, y, epochs=50)
```

```python
cgpa = float(input("Enter CGPA: "))
internships = int(input("Enter Internships: "))
projects = int(input("Enter Projects: "))
workshops = int(input("Enter Workshops/Certifications: "))
aptitude = int(input("Enter Aptitude Score: "))
softskills = float(input("Enter Soft Skills Rating: "))
extra = int(input("Extracurricular Activities (0/1): "))
training = int(input("Placement Training (0/1): "))
ssc = int(input("Enter SSC Marks: "))
hsc = int(input("Enter HSC Marks: "))
```

```python
test=np.array([[cgpa, internships, projects, workshops, aptitude, softskills, extra, training, ssc, hsc]])
test=scaler.transform(test)
pred=model.predict(test)
```

```python
if pred[0][0] > 0.5:
    print("Prediction:", le.inverse_transform([1])[0])
else:
    print("Prediction:", le.inverse_transform([0])[0])
```
